#### Subset the data to only include the first 5 columns

```python
df_sub = df.iloc[:,:8]
df_sub.head()
```

`.iloc[]` selects elements by their index. The first number/range selects rows the second columns. 

`.loc[]` works similarly but selects rows/columns by their name.

#### Drop the `Data.Health.Life Expectancy at Birth, Total` column

```python
df_sub.drop(df.columns[[7]], axis=1, inplace=True) #count from 0
df_sub.head()
```

`axis` specifies whether to drop rows `0` or columns `1`. `inplace` specifies whether to drop the elements in place or to reassign the dataframe.

You can also drop columns by name using `drop('column_name', axis=1, inplace=True)`

#### Sorting

```python
#by values
df_sub.sort_values(by="Year").head()
```

```python
#by axis
df_sub.sort_index(axis=1).head()
```

```python
#sort first by country, then by year
df_sub.sort_values(by=["Country","Year"]).head()
```

#### Calculations

```python
#Mean of a Column
meanBirth = df_sub.loc[:,["Data.Health.Birth Rate"]].mean()
meanBirth
```

```python
#Mean for each Country
meanBirthGrp = df_sub.loc[:,["Country", "Data.Health.Birth Rate"]].groupby(["Country"]).mean()
meanBirthGrp = meanBirthGrp.reset_index()
meanBirthGrp.sort_values(by=["Data.Health.Birth Rate"], ascending=False).head() 
```

```python
#Recalculate total life expectancy and add it as a column
df_sub["Life Expectancy at Birth, Total"] = (df_sub.iloc[:,5]+df_sub.iloc[:,6])/2
df_sub.head()
```

#### Selection

```python
#show only Countries with a mean birth rate below 15
meanBirthGrp.loc[meanBirthGrp["Data.Health.Birth Rate"]<15].sort_values(by=["Data.Health.Birth Rate"])
```

```python
#get all unique Country names
df_sub["Country"].unique()
```

```python
#Display only mean birth rates of five select countries
the5 = meanBirthGrp[["Country","Data.Health.Birth Rate"]].loc[meanBirthGrp["Country"].isin(
    ["United Kingdom","Nepal","Bulgaria","Mongolia","Comoros"])]
the5
```

```python
#Display the years in which Germany's birth rate was 10 or higher
df_sub["Year"].loc[(df_sub["Country"]=="Germany") & (df_sub["Data.Health.Birth Rate"]>=10)]
```

#### Plotting

```python
#Compare the change in birth rate over the years between germany and the worldwide average
import matplotlib.pyplot as plt
%matplotlib inline

#create dataframe with worldwide average

wwmeanBirth = df_sub.loc[:,["Year", "Data.Health.Birth Rate"]].groupby(["Year"]).mean().reset_index()

fig, ax = plt.subplots(1,1)
df_sub.loc[df_sub["Country"]=="Germany"].plot(x="Year",y="Data.Health.Birth Rate",ax=ax) 
#if x is unspecified it will be the index
#if y is unspecified it will plot every column
wwmeanBirth.plot(x="Year",y="Data.Health.Birth Rate",ax=ax)
ax.set_ylabel('Birth Rate')
ax.legend(["Germany","World Wide Average"])

plt.show()
```
