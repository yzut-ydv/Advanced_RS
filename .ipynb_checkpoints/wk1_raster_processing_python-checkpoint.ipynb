### Create a Raster stack

```python
from osgeo import gdal_array

# Load required rasters
# One as dataset to get Projection and Geotransform
data0 = gdal.Open("data\\LC08_L1TP_174037_20180904_20180912_01_T1\\LC08_L1TP_174037_20180904_20180912_01_T1_B2.TIF")

# Then all of the required ones as arrays
beirut_b2_arr = gdal_array.LoadFile("data\\LC08_L1TP_174037_20180904_20180912_01_T1\\LC08_L1TP_174037_20180904_20180912_01_T1_B2.TIF")
beirut_b3_arr = gdal_array.LoadFile("data\\LC08_L1TP_174037_20180904_20180912_01_T1\\LC08_L1TP_174037_20180904_20180912_01_T1_B3.TIF")
beirut_b4_arr = gdal_array.LoadFile("data\\LC08_L1TP_174037_20180904_20180912_01_T1\\LC08_L1TP_174037_20180904_20180912_01_T1_B4.TIF")
beirut_b5_arr = gdal_array.LoadFile("data\\LC08_L1TP_174037_20180904_20180912_01_T1\\LC08_L1TP_174037_20180904_20180912_01_T1_B5.TIF")

# Get Metadata from the dataset and size from one of the arrays 
y_pix, x_pix = beirut_b2_arr.shape
driver = gdal.GetDriverByName('GTiff')
geotrans = data0.GetGeoTransform()
proj = data0.GetProjection()

# Then write them to a new raster file
rOut = driver.Create("data\\results\\rgbnStack.tif", x_pix, y_pix, 4, gdal.GDT_UInt16)
rOut.GetRasterBand(1).WriteArray(beirut_b2_arr)
rOut.GetRasterBand(2).WriteArray(beirut_b3_arr)
rOut.GetRasterBand(3).WriteArray(beirut_b4_arr)
rOut.GetRasterBand(4).WriteArray(beirut_b5_arr)
rOut.SetGeoTransform(geotrans)
rOut.SetProjection(proj)
rOut.FlushCache()
rOut = None
```

```python
#Now load the new raster
rgbnStack = rasterio.open("data\\results\\rgbnStack.tif")

#check the structure of the stack
print(rgbnStack.profile)

#You can read a single band of the dataset into an array
#!Please note that rasterio and gdal count bands starting at 1, contrary to how python counts!
band2 = rgbnStack.read(1)
```

### Plot Band compositions

Normally a simple `show(rgbnStack.read([3,2,1]))` would be enough to plot the raster stack in true color.
In this case though you will first need to normalize the pixel values to a range of 0-255 and then cast from the datatype uint16 to uint8 to be able to plot it.

```python
#Function to normalize array values
def normalize(x, lower, upper):
    """Normalize an array to a given bound interval"""

    x_max = np.max(x)
    x_min = np.min(x)

    m = (upper - lower) / (x_max - x_min)
    x_norm = (m * (x - x_min)) + lower

    return x_norm

rgb_arr = rgbnStack.read([3,2,1])
nrg_arr = rgbnStack.read([4,3,2])
data_norm_rgb = np.array([normalize(rgb_arr[i,:,:], 0, 255) for i in range(rgb_arr.shape[0])])
data_norm_nrg = np.array([normalize(nrg_arr[i,:,:], 0, 255) for i in range(nrg_arr.shape[0])])
data_rgb = data_norm_rgb.astype("uint8")
data_nrg = data_norm_nrg.astype("uint8")

fig, ((axrgb, axnrg), (axrgbb, axnrgb)) = plt.subplots(2,2,figsize=(14,7))

show(data_rgb,title="True Color Composite", ax=axrgb)
show(data_nrg,title="False Color Composite", ax=axnrg)

#multiply the input data to increase or decrease brightness
show(data_rgb*2,title="True Color Composite brighter", ax=axrgbb)
show(data_nrg*2,title="False Color Composite brighter", ax=axnrgb)
plt.show()

#use `transform = rgbnStack.transform` for correct tick marks 
fig2, ax = plt.subplots(1, 1)
show(data_rgb*2,title="True Color Composite",transform=rgbnStack.transform,ax=ax)
ax.ticklabel_format(style = 'plain')
plt.show()
```

### Cropping

#### Cropping (Masking) via fiona and `rasterio.mask`

```python
import fiona
import rasterio.mask

with fiona.open("data/vector\\aoi_beirut.shp", "r") as shapefile:
    shapes = [feature["geometry"] for feature in shapefile] #reads all shapes in the shapefile (in this case 1)
    
with rasterio.open("data\\results\\rgbnStack.tif") as src:
    out_image, out_transform = rasterio.mask.mask(src, shapes, crop=True)
    out_meta = src.meta
    
out_meta.update({"driver": "GTiff",
                 "height": out_image.shape[1],
                 "width": out_image.shape[2],
                 "transform": out_transform})

with rasterio.open("data\\results\\rgbnCropRasterio.tif", "w", **out_meta) as dest:
    dest.write(out_image)

rgbnCropRasterio = rasterio.open("data\\results\\rgbnCropRasterio.tif")
```

#### Cropping via `gdal.Warp`

```python
gdal.Warp("data\\results\\rgbnCropGdal.tif", "data\\results\\rgbnStack.tif", 
          cutlineDSName="data/vector/aoi_beirut.shp", cropToCutline=True)

rgbnCropGdal = rasterio.open("data\\results\\rgbnCropGdal.tif")
```

Plot both for comparison

```python
fig3, (axrast, axgdal) = plt.subplots(1, 2, figsize=(14,7))
show(rgbnCropRasterio,title="Rasterio Crop",transform=rgbnCropRasterio.transform,ax=axrast)
show(rgbnCropGdal,title="GDAL Crop",transform=rgbnCropGdal.transform,ax=axgdal)
axrast.ticklabel_format(style = 'plain')
axgdal.ticklabel_format(style = 'plain')
```

As you can see both methods produce similar results, most likely even the same.

Please note that the black box surrounding the image is a result of how cropping in python works compared to how it works in R.
In R rasters are cropped using a bounding box surrounding the shape, python crops exactly to the shape and sets everything outside of it to 0.

#### Using rasterio

```python
from rasterio.enums import Resampling

# open band 10 with res 100m
TIRS = rasterio.open(os.path.join(landsat_loc, "LC08_L1TP_174037_20180904_20180912_01_T1_B10.TIF")) 

#NIR resolution is 30, TIRS resolution is 100 (or so we say, even though that is not the case). so we want NIR at 100 too.
#Thus the `upscale_factor` will be 3/10

upscale_factor = 3/10

# define path of band 5
NIR = os.path.join(landsat_loc, "LC08_L1TP_174037_20180904_20180912_01_T1_B5.TIF")

with rasterio.open(NIR) as dataset: # open band 5

    # resample data to target shape
    data = dataset.read(
        out_shape=(
            dataset.count,
            int(dataset.height * upscale_factor),
            int(dataset.width * upscale_factor)
        ),
        resampling=Resampling.bilinear
    )

    # scale image transform
    transform = dataset.transform * dataset.transform.scale(
        (dataset.width / data.shape[-1]),
        (dataset.height / data.shape[-2])
    )
    
    data_meta = dataset.meta
    
    print('Shape before resample:', dataset.shape)
    print('Shape after resample:', data.shape[1:])

data_meta.update({"driver": "GTiff",
                 "height": data.shape[1],
                 "width": data.shape[2],
                 "transform": transform})

# define path of band 5 after resampling
NIR_100m = os.path.join(landsat_loc, "band5_resampled.tif")

# export resampled band
with rasterio.open(NIR_100m, "w", **data_meta) as dest:
    dest.write(data)
```

### Panchromatic sharpening

#### In Python

Please note that the processing can take some time. RuntimeWarnings may also accur, but are of no consequence.

```python
'''
@author: Gerasimos Michalitsianos
Science Systems and Applications, Inc. 
June 2015
https://github.com/hustxujinkang/image-pan-sharpening
'''

#Function for writing the images

def writeImg(imgs, outname, dspan):

    nrows,ncols= imgs[0].shape 
    driv = gdal.GetDriverByName('GTiff')
    dst = driv.Create(outname, ncols, nrows, 4, gdal.GDT_Float32)
    dst.SetGeoTransform(dspan.GetGeoTransform())
    dst.SetProjection(dspan.GetProjection())
    dst.GetRasterBand(1).WriteArray(imgs[2])
    dst.GetRasterBand(2).WriteArray(imgs[1])
    dst.GetRasterBand(3).WriteArray(imgs[0])
    dst.GetRasterBand(4).WriteArray(imgs[3])
    dst=None
    del dst

#Crop the panchromatic band

gdal.Warp("data\\results\\panCrop.tif", 
          "data\\LC08_L1TP_174037_20180904_20180912_01_T1\\LC08_L1TP_174037_20180904_20180912_01_T1_B8.tif", 
          cutlineDSName="data/vector\\aoi_beirut.shp", cropToCutline=True)    
    
#Resample rgbnStack to match the panchromatic band

multids = gdal.Open("data\\results\\rgbnCropGdal.tif")
dspan = gdal.Open("data\\results\\panCrop.tif")
outname = "data\\results\\rgbnResamp.tif"

srcProjection     = multids.GetProjection()
srcGeotransform   = multids.GetGeoTransform()
matchProjection   = dspan.GetProjection()
matchGeotransform = dspan.GetGeoTransform()
ncols             = dspan.RasterXSize
nrows             = dspan.RasterYSize


dst_ds = gdal.GetDriverByName('GTiff').Create(outname, ncols, nrows, 4, gdal.GDT_UInt16)
dst_ds.SetGeoTransform(matchGeotransform)
dst_ds.SetProjection(matchProjection)
gdal.ReprojectImage(multids, dst_ds, srcProjection, matchProjection, gdal.GRA_Cubic)
dst_ds = None 
del dst_ds

#broveySharpen

dsmulti = gdal.Open(outname)

pan   = dspan.GetRasterBand(1).ReadAsArray().astype(float)
red   = dsmulti.GetRasterBand(3).ReadAsArray() 
green = dsmulti.GetRasterBand(2).ReadAsArray()
blue  = dsmulti.GetRasterBand(1).ReadAsArray()
nir   = dsmulti.GetRasterBand(4).ReadAsArray()

redbrovsharp   = np.multiply( np.true_divide(red , red + green + blue + nir), pan)
greenbrovsharp = np.multiply( np.true_divide(green , red + green + blue + nir), pan)
bluebrovsharp  = np.multiply( np.true_divide(blue , red + green + blue + nir), pan)
nirbrovsharp   = np.multiply( np.true_divide(nir , red + green + blue + nir), pan)

brovbands = [redbrovsharp,greenbrovsharp,bluebrovsharp,nirbrovsharp]

#fihsSharpen

pan   = dspan.GetRasterBand(1).ReadAsArray().astype(float)
red   = dsmulti.GetRasterBand(3).ReadAsArray() 
green = dsmulti.GetRasterBand(2).ReadAsArray()
blue  = dsmulti.GetRasterBand(1).ReadAsArray()
nir   = dsmulti.GetRasterBand(4).ReadAsArray()

L = (red+green+blue+nir)/float(4)
redfihssharp   = red   + (pan - L) 
greenfihssharp = green + (pan - L)
bluefihssharp  = blue  + (pan - L)
nirfihssharp   = nir   + (pan - L) 

fihsbands = [redfihssharp,greenfihssharp,bluefihssharp,nirfihssharp]

#write

writeImg(brovbands, "data\\results\\panSharpen_brovey.tif", dspan)
writeImg(fihsbands, "data\\results\panSharpen_fihs.tif", dspan)
```

### Mosaicking

```python
img1 = "data\\NTL\\VNP46A2.A2019100.h26v06.001.2021020141752_DNB_BRDF-Corrected_NTL.tif"
img2 = "data\\NTL\\VNP46A2.A2019100.h26v07.001.2021020141743_DNB_BRDF-Corrected_NTL.tif"
outname = "data\\results\\Mosaic.tif"

imgList = [img1, img2]

vrt = gdal.BuildVRT("mos.vrt", imgList)
gdal.Translate(outname, vrt)

r1 = rasterio.open(img1)
r2 = rasterio.open(img2)
mos = rasterio.open(outname)

fig4, (ax1, ax2, axmos) = plt.subplots(1, 3, figsize=(14,7))
show(r1,title="Raster 1",ax=ax1)
show(r2,title="Raster 2",ax=ax2)
show(mos,title="Mosaic",ax=axmos)
ax1.ticklabel_format(style = 'plain')
ax2.ticklabel_format(style = 'plain')
axmos.ticklabel_format(style = 'plain')
```
